# Visualisation of Raw Data

Here we will visualise the raw data.

### Import libraries:

(Note: this may take a while.)

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("../style_Matplotlib_charts.mplstyle")
```

### Load the file

Load the external `.csv` file, and load into local variable.

```python
data_frame = pd.read_csv('raw_data.csv', delimiter = ',')
```

### Info about data

Includes:
 - Column
 - Non-null Count
 - Data-type

```python
data_frame.info()
```

### Statistics

Includes:
 - Count
 - Mean
 - Standard deviation
 - Minimum
 - 25th percentile
 - Median
 - 75th percentile
 - Maximum

```python
data_frame = data_frame.drop(columns=['wmo', 'lat', 'lon'])
data_frame.describe()
```

```python
# Calculate skewness
skewness = data_frame.skew()

print(f"Skewness:\n{skewness}")
```

### Graphing data

Histogram:

```python
print(data_frame.columns)
```

```python
x_axis = "air_temp"
if x_axis == 'wind_dir_deg':
    data_frame[x_axis].hist(bins=72)
else:
    data_frame[x_axis].hist(bins=30)
plt.xlabel(x_axis)
plt.ylabel('Frequency')
plt.title(f"Histogram of {x_axis}")
plt.show()
```

Line Chart:

```python
data_frame["air_temp_moving_avg"] = data_frame["air_temp"].rolling(window=100).mean()
data_frame.plot(
    x="time", y="air_temp", kind="line", ax=plt.gca(), label="Air Temperature"
)

data_frame.plot(
    x="time",
    y="air_temp_moving_avg",
    kind="line",
    ax=plt.gca(),
    label="Moving Average (1000)",
)


plt.xlabel('Index')
plt.ylabel('Air Temperature')
plt.title('Line Graph of Air Temperature')
plt.show()
```
