# Feature Engineering
## Load required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('wrangled_data.csv')

columns = ["air_temp", "apparent_t"]

# Plotting the data vs each other
plt.style.use("../../style_Matplotlib_charts.mplstyle")
plt.figure(figsize=(9, 9))

plt.scatter(
    data_frame[columns[0]],
    data_frame[columns[1]],
    label=f"{columns[0]} vs {columns[1]}",
    s=20,
    marker="x",
)
plt.xlabel(columns[0])
plt.ylabel(columns[1])
plt.title(f"{columns[0]} vs {columns[1]}")
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.grid()
plt.legend()
plt.show()
```
