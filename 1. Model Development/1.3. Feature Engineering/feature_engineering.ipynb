# Feature Engineering
## Load required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('wrangled_data.csv')
```

```python
data_frame["gust_diff"] = data_frame["gust_kmh"] - data_frame["wind_spd_kmh"]
# Plotting the data vs each other
plt.figure(figsize=(9, 9))
plt.scatter(
    data_frame["wind_spd_kmh"],
    data_frame["gust_diff"],
    label=f"wind_speed vs gust-diff",
    s=20,
    marker="x",
)
plt.xlabel("wind_spd_kmh")
plt.ylabel('gust-diff')
plt.title(f"wind_speed vs gust-diff")
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.grid()
```

```python
# save new features to csv
data_frame.to_csv("../1.4. Model Training/1.4.1.model_ready_data.csv", index=False)
```
